### DataFrame의 요소 접근하기 : 읽기와 수정

## 통계 기초
평균 mean : (데이터들의 합) / (데이터 수) 중심성향에 대한 추정량을 계산<br>
편차 deviation : 관측값과 평균갑의 차이 (데이터값 - 평균값)<br>
분산 variance : 편차의 제곱의 합의 평균, 음수값이 양수로 더해진다.<br>
표준편차 standard deviation : 분산에 루트(제급근)을 씌운 값<br>

```python
df.mad() # 평균 절대 편차
```

```python
df.mean() # 평균
```

```python
df.median() # 중위수, 2사분위수
```

```python
df.sum() # 합
df.max()
df.min()
df.mode() # 최빈값, 빈도가 가장 높은 값
df.var() # 분산
df.skew() # 비대칭도
df.kurt() # 첨도
```

## 데이터 프레임 추가와 합치기

```python
# 행(수직)으로 합치기 : append(), concat()
score_table1 = {'영어' : [10, 20, 30, 40 ,50],
               '수학' : [50, 40, 30, 20, 10]}
score_table2 = {'영어' : [60, 70, 80, 90 ,100],
               '수학' : [100, 90, 80, 70, 60]}

df1 = pd.DataFrame(score_table1)
df2 = pd.DataFrame(score_table2)
new_df1 = df1.append(df2)
new_df1
new_df2 = df1.append(df2, ignore_index = True) # 인덱스 중복을 없앰
new_df2
new_df3 = new_df2.append(new_df2, ignore_index = True) # 인덱스 중복을 없앰
new_df3
```

```python
pd.concat([df1, df2], axis = 0, ignore_index = True)
```

```python
# 열(수평) 합치기 : concat(), merge(), join()
pd.concat([df1, df2], axis = 1)
```

```python
# merge()
dests = pd.read_csv('dest.csv')
tips = pd.read_csv('tips.csv')
print(dests)
print(tips)
# inner join 공통조건이있을경우 조인된다.
pd.merge(dests, tips, how = 'inner', on = 'EmpNr')
```

```python
# outer join : 공통조건 관계없이 join된다. 결측치는 Nan으로 표기
pd.merge(dests, tips, how='outer')
```

```python
# join(other, on = None, how = 'left', lsuffix='', rsuffix='', sor)
dests.join(tips, lsuffix='_Dest', rsuffix='_Tips') # Left join 
```
