1. pandas로 boston_train.csv 파일을  읽어서

  1.1 속성 출력: shape,index,value

  1.2 행(row)으로 접근

  1.3 열(column)로 접근

  1.4 열의 이름을 직접 사용하여 열을 추출하여 Series 객체 생성

  1.5 열의 번호를 사용하여 열을 추출하여 Series 객체 생성

  1.6 열의 이름을 직접 사용하여 여러개 열을 추출하여
      DataFrame 객체 생성  
 
  1.7 열의 번호를 사용하여 여러개 열을 추출하여
      DataFrame 객체 생성  

  1.8 df.iloc[행,열]를 사용하여 스칼라 값(원소) 값을 접근하고 수정
  
  1.9 수정된 DataFrame을 새로운 csv 파일로 저장

```python
import pandas as pd

train = pd.read_csv('../boston_train.csv')
train
```

```python
print('1.1 속성 출력: shape,index,value')
print('shape :', train.shape)
print('index :', train.index)
print('values :', train.values)
```

```python
print('1.2 행(row)으로 접근')
train[0:]
```

```python
print('1.3 열(column)로 접근')
train[train.columns]
```

```python
print('1.4 열의 이름을 직접 사용하여 열을 추출하여 Series 객체 생성')
trainSeries = train['CRIM']
print(trainSeries)
print(trainSeries.shape)
print(type(trainSeries))
```

```python
print('1.5 열의 번호를 사용하여 열을 추출하여 Series 객체 생성')
trainSeries2 = train[train.columns[0]]
print(trainSeries2)
print(trainSeries2.shape)
print(type(trainSeries2))
```

```python
print('1.6 열의 이름을 직접 사용하여 여러개 열을 추출하여 DataFrame 객체 생성')
trainDF1 = train[['CRIM', 'ZN', 'INDUS', 'NOX']]
print(trainDF1)
print(trainDF1.shape)
print(type(trainDF1))
```

```python
print('1.7 열의 번호를 사용하여 여러개 열을 추출하여 DataFrame 객체 생성')
trainDF2 = train[[train.columns[0], train.columns[1], train.columns[2]]]
print(trainDF2)
print(trainDF2.shape)
print(type(trainDF2))
```

```python
print('1.8 df.iloc[행,열]를 사용하여 스칼라 값(원소) 값을 접근하고 수정')
print(train.iloc[0, 0])
train.iloc[0, 0] = 5.0000
print(train.iloc[0, 0])
train.iloc[0, 0] = 7.123
print(train.iloc[0, 0])
```

```python
print('1.9 수정된 DataFrame을 새로운 csv 파일로 저장')
```

```python
train.to_csv('../boston_train_modify.csv', index = False)
train_modify = pd.read_csv('../boston_train_modify.csv')
print(train_modify)
print(train_modify.iloc[0, 0])
```
